# PARTIE 1 — Les opérateurs de base

Python peut s'utiliser comme une calculatrice scientifique. Voici tous les opérateurs mathématiques.

## 1.1 — `+`  `-`  `*`  `/` : les quatre opérations

````python
# + addition
# Exemple : masse totale de deux réactifs
masse_glucose   = 18.0   # g
masse_eau       = 36.0   # g
masse_totale    = masse_glucose + masse_eau
print("Masse totale :", masse_totale, "g")
````

````python
# - soustraction
# Exemple : volume restant dans un flacon
volume_initial  = 500   # mL
volume_preleve  = 125   # mL
volume_restant  = volume_initial - volume_preleve
print("Volume restant :", volume_restant, "mL")
````

````python
# * multiplication
# Exemple : calcul de dose (mg/kg × kg)
dose_par_kg = 5      # mg/kg
poids       = 70     # kg
dose        = dose_par_kg * poids
print("Dose à administrer :", dose, "mg")
````

````python
# / division
# Exemple : concentration après dissolution
masse_soluté = 5.0    # g
volume_L     = 0.5    # L
concentration = masse_soluté / volume_L
print("Concentration :", concentration, "g/L")
````

## 1.2 — `**` : la puissance

`**` signifie *"élevé à la puissance"*. C'est l'équivalent de l'exposant en mathématiques.

```
2 ** 3  =  2³  =  8
10 ** 6 =  10⁶ =  1 000 000
```

````python
# Exemple simple
print(2 ** 3)    # 2³ = 8
print(10 ** 6)   # 10⁶ = 1 000 000
print(2 ** 0.5)  # racine carrée de 2
````

````python
# Application : doublement bactérien
# Après t heures, N = N0 × 2^t
N0 = 100    # bactéries au départ
t  = 5      # heures

N = N0 * (2 ** t)
print("Après", t, "heures :", N, "bactéries")
````

````python
# Application : loi de Beer-Lambert  A = ε × l × c
epsilon = 6500    # L/mol/cm
l       = 1       # cm
c       = 0.002   # mol/L

A = epsilon * l * c
print("Absorbance :", A)
````

## 1.3 — `//` et `%` : division entière et reste

| Opérateur | Nom | Exemple | Résultat |
|---|---|---|---|
| `//` | division entière | `17 // 5` | `3` |
| `%` | modulo (reste) | `17 % 5` | `2` |

> `17 = 5 × 3 + 2`  →  quotient = 3, reste = 2

````python
# Exemple : répartir 17 échantillons dans des boîtes de 5
echantillons = 17
capacite     = 5

boites_pleines = echantillons // capacite
reste          = echantillons % capacite

print("Boîtes pleines :", boites_pleines)
print("Échantillons restants :", reste)
````

## 1.4 — Tableau récapitulatif des opérateurs

| Opérateur | Nom | Exemple | Résultat |
|---|---|---|---|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Soustraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division | `5 / 2` | `2.5` |
| `**` | Puissance | `2 ** 3` | `8` |
| `//` | Division entière | `5 // 2` | `2` |
| `%` | Modulo (reste) | `5 % 2` | `1` |

````python
# Teste tous les opérateurs ici
a = 10
b = 3

print("a + b  =", a + b)
print("a - b  =", a - b)
print("a * b  =", a * b)
print("a / b  =", a / b)
print("a ** b =", a ** b)
print("a // b =", a // b)
print("a % b  =", a % b)
````

# PARTIE 2 — Les fonctions intégrées

Python fournit des **fonctions prêtes à l'emploi** — pas besoin de les programmer, il suffit de les appeler.

Une fonction s'utilise toujours ainsi : `nom_fonction(valeur)`

## 2.1 — `len()` : compter les éléments

`len()` retourne le **nombre d'éléments** dans une liste (ou le nombre de caractères dans un texte).

```
len( [10, 20, 30, 40] )  →  4
         ↑   ↑   ↑   ↑
         1   2   3   4   →  4 éléments
```

````python
# len() sur une liste de nombres
absorbances = [0.45, 0.52, 0.48, 0.61, 0.39]

nombre = len(absorbances)
print("Nombre de mesures :", nombre)
````

````python
# len() sur une liste de noms
patients = ["Ali", "Sara", "Karim", "Lina"]

print("Nombre de patients :", len(patients))
````

````python
# Utiliser len() dans un calcul : moyenne manuelle
mesures = [0.452, 0.448, 0.461, 0.455, 0.450]

total   = 0.452 + 0.448 + 0.461 + 0.455 + 0.450
moyenne = total / len(mesures)   # len() donne 5

print("Total :", total)
print("Nombre de mesures :", len(mesures))
print("Moyenne :", moyenne)
````

## 2.2 — `sum()` : additionner tous les éléments

`sum()` additionne **tous les éléments** d'une liste en une seule instruction.

````python
# Avant sum() : on additionne à la main
notes = [12, 15, 9, 18, 11]

total_manuel = 12 + 15 + 9 + 18 + 11
print("Total manuel :", total_manuel)
````

````python
# Avec sum() : une seule ligne, quelle que soit la taille de la liste
notes = [12, 15, 9, 18, 11]

total = sum(notes)
print("Total avec sum() :", total)
````

````python
# Combinaison sum() + len() → moyenne
absorbances = [0.45, 0.52, 0.48, 0.61, 0.39]

moyenne = sum(absorbances) / len(absorbances)
print("Moyenne des absorbances :", moyenne)
````

## 2.3 — `min()` et `max()` : le minimum et le maximum

````python
# Trouver les valeurs extrêmes d'une série de mesures
températures = [36.5, 37.2, 38.9, 36.8, 39.5, 37.0]

print("Température minimale :", min(températures), "°C")
print("Température maximale :", max(températures), "°C")
````

````python
# Application : identifier la plage de variation
diamètres = [18, 22, 25, 19, 21, 28, 14]

valeur_min  = min(diamètres)
valeur_max  = max(diamètres)
etendue     = valeur_max - valeur_min

print("Diamètre minimum :", valeur_min, "mm")
print("Diamètre maximum :", valeur_max, "mm")
print("Étendue           :", etendue, "mm")
````

## 2.4 — `round()` : arrondir un nombre

`round(valeur, n)` arrondit à `n` décimales.

````python
# Sans arrondi : résultat brut de Python
moyenne = 7 / 3
print("Sans arrondi :", moyenne)
````

````python
# Avec round()
moyenne = 7 / 3

print("round(moyenne, 0) :", round(moyenne, 0))   # entier
print("round(moyenne, 1) :", round(moyenne, 1))   # 1 décimale
print("round(moyenne, 2) :", round(moyenne, 2))   # 2 décimales
print("round(moyenne, 3) :", round(moyenne, 3))   # 3 décimales
````

# PARTIE 3 — Statistiques avec `statistics`

Python possède un module intégré appelé `statistics` qui fournit les fonctions statistiques essentielles.

Pour l'utiliser, il faut d'abord l'**importer** avec `import statistics`.

## 3.1 — Qu'est-ce qu'un `import` ?

Python de base ne charge pas tout en mémoire au démarrage. Les fonctions spécialisées sont rangées dans des **modules** qu'on charge à la demande.

```python
import statistics   # charge le module
statistics.mean()   # utilise une fonction du module
```

C'est comme ouvrir une boîte à outils spécialisée avant de s'en servir.

````python
import statistics   # à exécuter UNE FOIS avant les cellules suivantes
````

## 3.2 — `mean()` : la moyenne

La moyenne = somme de tous les éléments ÷ nombre d'éléments.

````python
# Moyenne manuelle vs statistics.mean()
absorbances = [0.45, 0.52, 0.48, 0.61, 0.39]

# Méthode manuelle
moyenne_manuelle = sum(absorbances) / len(absorbances)

# Avec statistics
moyenne_auto = statistics.mean(absorbances)

print("Moyenne manuelle :", moyenne_manuelle)
print("Moyenne auto     :", moyenne_auto)
````

## 3.3 — `median()` : la médiane

La médiane est la **valeur centrale** quand on trie les données. Elle est moins sensible aux valeurs extrêmes que la moyenne.

````python
# Diamètres d'inhibition (mm)
diamètres = [14, 18, 19, 21, 22, 25, 28]
#             ↑                       ↑
#           valeur                  valeur
#           min                     max
#                        ↑
#                     médiane = 21 (valeur du milieu)

print("Médiane :", statistics.median(diamètres), "mm")
print("Moyenne :", statistics.mean(diamètres), "mm")
````

````python
# Pourquoi la médiane est utile : avec une valeur aberrante
doses = [200, 210, 195, 205, 198, 950]   # 950 = erreur de saisie ?

print("Moyenne  :", round(statistics.mean(doses), 1), "mg")    # faussée par 950
print("Médiane  :", statistics.median(doses), "mg")            # pas affectée
````

## 3.4 — `variance()` et `stdev()` : variance et écart-type

- La **variance** mesure la dispersion des valeurs autour de la moyenne
- L'**écart-type** (stdev = standard deviation) est la racine carrée de la variance — plus facile à interpréter car dans la même unité que les données

> Un écart-type faible = mesures très reproductibles. Un écart-type élevé = grande variabilité.

````python
# Réplicats d'une mesure de DO
DO = [0.452, 0.448, 0.461, 0.455, 0.450]

print("Moyenne   :", round(statistics.mean(DO), 4))
print("Variance  :", round(statistics.variance(DO), 6))
print("Écart-type:", round(statistics.stdev(DO), 4))
````

````python
# Comparer deux séries : même moyenne, dispersions différentes
serie_A = [48, 50, 52, 49, 51]   # résultats groupés
serie_B = [30, 45, 55, 65, 55]   # résultats dispersés

print("--- Série A ---")
print("Moyenne    :", statistics.mean(serie_A))
print("Écart-type :", round(statistics.stdev(serie_A), 2))

print("\n--- Série B ---")
print("Moyenne    :", statistics.mean(serie_B))
print("Écart-type :", round(statistics.stdev(serie_B), 2))
````

# PARTIE 4 — Le problème de la répétition

On sait maintenant calculer des statistiques sur **une** liste. Mais que se passe-t-il quand on a **plusieurs** séries de données ?

````python
# On a 4 patients — on veut calculer la dose pour chacun
# Sans boucle : on répète le même calcul 4 fois

dose_kg = 5   # mg/kg

dose_Ali   = 70 * dose_kg
dose_Sara  = 55 * dose_kg
dose_Karim = 90 * dose_kg
dose_Lina  = 48 * dose_kg

print("Ali   :", dose_Ali,   "mg")
print("Sara  :", dose_Sara,  "mg")
print("Karim :", dose_Karim, "mg")
print("Lina  :", dose_Lina,  "mg")
````

Ce code fonctionne... mais :
- Si on a **100 patients** → 100 lignes identiques
- Si on veut **changer la formule** → modifier 100 lignes
- Si on **ajoute un patient** → réécrire du code

**La solution : regrouper les données dans une liste, puis parcourir la liste avec une boucle `for`.**

# PARTIE 5 — La boucle `for`

La boucle `for` dit à Python : *"pour chaque élément de cette liste, exécute ce bloc de code"*.

```python
for variable in séquence:
    # code répété à chaque tour
```

## 5.1 — Premier exemple : le plus simple possible

````python
# Afficher chaque élément d'une liste
absorbances = [0.45, 0.52, 0.48, 0.61, 0.39]

for A in absorbances:
    print(A)
````

````python
# Faire un calcul sur chaque élément
# Beer-Lambert : c = A / (ε × l)
absorbances = [0.45, 0.52, 0.48, 0.61, 0.39]
epsilon = 6500
l = 1

for A in absorbances:
    c = A / (epsilon * l)
    print(c)
````

## 5.2 — `range()` : générer une séquence de nombres

| Écriture | Résultat |
|---|---|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 11, 2)` | 0, 2, 4, 6, 8, 10 |

> ⚠️ La borne de fin est toujours **exclue**.

````python
# Numéroter des tubes
for i in range(1, 6):
    print("Tube", i)
````

````python
# Croissance bactérienne heure par heure
N0 = 100

for t in range(0, 9):          # de t=0 à t=8
    N = N0 * (2 ** t)          # ** = puissance
    print("t =", t, "h  →  N =", N, "bactéries")
````

## 5.3 — Combiner `for` + `len()` + `range()`

Quand on a **deux listes** à parcourir en parallèle, on utilise `range(len(liste))`.

````python
# Deux listes en parallèle : patients et poids
patients = ["Ali", "Sara", "Karim", "Lina"]
poids    = [70,    55,     90,      48]
dose_kg  = 5

# range(len(patients)) génère : 0, 1, 2, 3
# i sert d'indice pour accéder aux deux listes simultanément

for i in range(len(patients)):
    dose = poids[i] * dose_kg
    print(patients[i], "→", dose, "mg")
````

## 5.4 — Accumuler avec `for` : reconstruire `sum()` à la main

Comprendre comment `sum()` fonctionne en interne :

````python
mesures = [0.452, 0.448, 0.461, 0.455, 0.450]

total = 0                  # accumulateur — commence à zéro

for valeur in mesures:
    total = total + valeur  # on ajoute chaque valeur au total
    print("Après ajout de", valeur, "→ total =", total)

moyenne = total / len(mesures)
print("\nMoyenne finale :", moyenne)
````

# PARTIE 6 — Les f-strings : afficher proprement

Jusqu'ici on a utilisé `print("texte", variable)`. C'est fonctionnel, mais il existe une méthode plus élégante et plus puissante : les **f-strings**.

## 6.1 — Syntaxe de base

```python
f"texte {variable} texte"
↑
ce f active les { }
```

Le `f` devant les guillemets dit à Python : *"les accolades `{ }` contiennent des variables — remplace-les par leur valeur"*.

````python
# Comparaison : ancien style vs f-string
nom  = "Sara"
dose = 275

# Ancien style
print("Patient :", nom, "→ Dose :", dose, "mg")

# Avec f-string
print(f"Patient : {nom} → Dose : {dose} mg")
````

````python
# On peut même faire des calculs DANS les { }
poids   = 55
dose_kg = 5

print(f"Dose pour {poids} kg : {poids * dose_kg} mg")
````

## 6.2 — Formater les nombres : `:.nf`

```
{ variable : .3f }
              ↑↑↑
              ||└─ f = float (nombre décimal)
              |└── 3 = 3 chiffres après la virgule
              └─── : = début du format
```

````python
# Sans format : résultat brut de Python
moyenne = sum([0.452, 0.448, 0.461, 0.455, 0.450]) / 5
print(f"Sans format : {moyenne}")
print(f"Avec .1f    : {moyenne:.1f}")
print(f"Avec .2f    : {moyenne:.2f}")
print(f"Avec .3f    : {moyenne:.3f}")
print(f"Avec .4f    : {moyenne:.4f}")
````

````python
# Notation scientifique avec :e
N = 25600000   # bactéries

print(f"Notation standard    : {N}")
print(f"Notation scientifique: {N:.2e}")
````

## 6.3 — f-strings dans une boucle `for` : la combinaison parfaite

C'est là que tout se connecte. On utilise la boucle `for` pour parcourir les données, et les f-strings pour afficher les résultats proprement.

````python
# Gamme étalon — Beer-Lambert avec f-string
epsilon       = 6500
l             = 1
concentrations = [0.001, 0.002, 0.005, 0.010]

print("--- Gamme étalon ---")
for c in concentrations:
    A = epsilon * l * c
    print(f"c = {c:.3f} mol/L  →  Absorbance = {A:.3f}")
````

````python
# Calcul de dose avec f-string — le code du notebook original
patients = ["Ali", "Sara", "Karim", "Lina"]
poids    = [70, 55, 90, 48]
dose_kg  = 5
dose_max = 400

print("--- Calcul des doses ---")
for i in range(len(patients)):
    dose = poids[i] * dose_kg
    if dose > dose_max:
        print(f"⚠️  {patients[i]} ({poids[i]} kg) : dose TROP ÉLEVÉE ({dose} mg)")
    else:
        print(f"✅  {patients[i]} ({poids[i]} kg) : dose correcte ({dose} mg)")
````

````python
# Rapport statistique complet avec f-strings
DO = [0.452, 0.448, 0.461, 0.455, 0.450]

n      = len(DO)
mini   = min(DO)
maxi   = max(DO)
moy    = statistics.mean(DO)
med    = statistics.median(DO)
ectype = statistics.stdev(DO)

print("==============================")
print("  Rapport statistique — DO")
print("==============================")
print(f"  N          : {n}")
print(f"  Minimum    : {mini:.3f}")
print(f"  Maximum    : {maxi:.3f}")
print(f"  Moyenne    : {moy:.3f}")
print(f"  Médiane    : {med:.3f}")
print(f"  Écart-type : {ectype:.4f}")
print("==============================")
````
