## 1. Pourquoi a-t-on besoin des conditions ?

En laboratoire, tu interprètes constamment des résultats selon des seuils :

- Si la glycémie est > 1.10 g/L → hyperglycémie
- Si le diamètre d'inhibition est < 20 mm → résistant
- Si la dose dépasse 400 mg → risque toxique

Sans conditions, Python exécuterait **toujours** le même code.  
Les conditions permettent à Python de **choisir quoi faire** selon la valeur d'une variable.

````python
# Sans condition : Python affiche toujours la même chose
absorbance = 1.20
print("Absorbance :", absorbance)
# → ne dit rien sur la validité du résultat
````

````python
# Avec condition : Python interprète le résultat
absorbance = 1.20

if absorbance > 0.9:
    print("⚠️  Hors gamme — diluer l'échantillon")
````

## 2. Le `if` seul — "si"

```python
if condition:
    # code exécuté UNIQUEMENT si la condition est vraie
```

| Élément | Rôle |
|---|---|
| `if` | Mot-clé : signifie **"si"** |
| `condition` | Expression vraie (`True`) ou fausse (`False`) |
| `:` | Obligatoire — annonce le bloc conditionnel |
| indentation | 4 espaces — ce qui est décalé s'exécute si condition vraie |

> Si la condition est `False` → le bloc est **complètement ignoré**.

````python
# Condition VRAIE → le bloc s'exécute
temperature = 38.9

if temperature > 37.5:
    print("Fièvre détectée :", temperature, "°C")   # ← exécuté

print("Fin de la vérification")   # toujours affiché
````

````python
# Condition FAUSSE → le bloc est ignoré
temperature = 36.8

if temperature > 37.5:
    print("Fièvre détectée")      # ← ignoré car 36.8 > 37.5 est FAUX

print("Fin de la vérification")   # toujours affiché
````

### Ce qui se passe dans Python :

```
temperature = 38.9
        ↓
if temperature > 37.5 ?
        ↓
    38.9 > 37.5  →  TRUE
        ↓
print("Fièvre détectée")   ← exécuté
```

```
temperature = 36.8
        ↓
if temperature > 37.5 ?
        ↓
    36.8 > 37.5  →  FALSE
        ↓
print("Fièvre détectée")   ← IGNORÉ
```

## 3. Le `if / else` — "si... sinon"

Avec `if` seul, si la condition est fausse → rien ne se passe.  
Avec `else`, on dit à Python **quoi faire dans l'autre cas**.

```python
if condition:
    # si la condition est VRAIE
else:
    # si la condition est FAUSSE
```

> L'un **ou** l'autre bloc s'exécute toujours — jamais les deux en même temps.

````python
# Exemple simple : résultat positif ou négatif
anticorps = 1.2    # unités arbitraires
seuil     = 1.0

if anticorps >= seuil:
    print("✅ Résultat POSITIF")
else:
    print("❌ Résultat NÉGATIF")
````

````python
# Modifie la valeur de anticorps et réexécute pour voir l'autre cas
anticorps = 0.6
seuil     = 1.0

if anticorps >= seuil:
    print("✅ Résultat POSITIF")
else:
    print("❌ Résultat NÉGATIF")
````

````python
# Application pharmacologie : dose toxique ou non
poids    = 90    # kg
dose_kg  = 5     # mg/kg
dose_max = 400   # seuil toxique

dose = poids * dose_kg

if dose > dose_max:
    print(f"⚠️  Dose TROP ÉLEVÉE : {dose} mg — réduire la prescription")
else:
    print(f"✅  Dose correcte : {dose} mg")
````

## 4. Le `if / elif / else` — "si... sinon si... sinon"

Quand il y a **plus de deux cas possibles**, on ajoute des `elif`.

`elif` est la contraction de **else if** — *"sinon si"*.

```python
if condition_1:
    # cas 1
elif condition_2:
    # cas 2
elif condition_3:
    # cas 3
else:
    # tous les autres cas
```

> Python teste les conditions **dans l'ordre**, et s'arrête dès qu'une est vraie.  
> Le `else` final est **facultatif** mais conseillé pour capturer les cas imprévus.

````python
# Classification du pH — 3 cas possibles
pH = 5.2

if pH < 7.0:
    print(f"pH = {pH} → Milieu ACIDE")
elif pH == 7.0:
    print(f"pH = {pH} → Milieu NEUTRE")
else:
    print(f"pH = {pH} → Milieu BASIQUE")
````

````python
# Classification de la fièvre — 4 cas
# Modifie la valeur pour tester chaque cas
temperature = 39.2

if temperature < 37.5:
    print(f"{temperature}°C → Normal")
elif temperature < 38.5:
    print(f"{temperature}°C → Fièvre légère")
elif temperature < 40.0:
    print(f"{temperature}°C → Fièvre modérée")
else:
    print(f"{temperature}°C → ⚠️  Hyperthermie sévère")
````

### Comment Python lit ce code :

```
temperature = 39.2
      ↓
temperature < 37.5 ?  →  NON
      ↓
temperature < 38.5 ?  →  NON
      ↓
temperature < 40.0 ?  →  OUI
      ↓
"Fièvre modérée"   ← exécuté, Python s'arrête ici
```

> Dès qu'une condition est vraie, Python **n'évalue plus les suivantes**.

## 5. Les opérateurs de comparaison

| Opérateur | Signification | Exemple | Résultat |
|---|---|---|---|
| `==` | égal à | `pH == 7.0` | True si pH vaut exactement 7.0 |
| `!=` | différent de | `résultat != "négatif"` | True si différent |
| `<` | strictement inférieur | `A < 0.1` | True si A inférieur à 0.1 |
| `<=` | inférieur ou égal | `dose <= 400` | True si dose ≤ 400 |
| `>` | strictement supérieur | `N > 10000` | True si N supérieur à 10000 |
| `>=` | supérieur ou égal | `diamètre >= 20` | True si diamètre ≥ 20 |

> ⚠️ Ne jamais confondre `=` (donner une valeur) et `==` (comparer)

````python
# Voir True et False directement
c = 0.5

print(c == 0.5)   # True
print(c != 1.0)   # True
print(c > 1.0)    # False
print(c <= 0.5)   # True
print(c < 0.5)    # False
````

## 6. Combiner des conditions : `and`, `or`, `not`

| Opérateur | Signification | Résultat |
|---|---|---|
| `and` | ET — les **deux** conditions doivent être vraies | `True and True → True` |
| `or` | OU — **au moins une** condition vraie suffit | `True or False → True` |
| `not` | NON — **inverse** la condition | `not True → False` |

````python
# and : zone thérapeutique
# la dose doit être ENTRE 50 et 400 mg
dose = 280

if dose >= 50 and dose <= 400:
    print(f"✅  {dose} mg : dans la zone thérapeutique")
else:
    print(f"⚠️  {dose} mg : hors zone thérapeutique")
````

````python
# or : alerte pH extrême
# danger si pH trop acide OU trop basique
pH = 3.5

if pH < 4.0 or pH > 9.0:
    print(f"🚨  pH = {pH} : conditions létales pour les cellules")
else:
    print(f"pH = {pH} : conditions acceptables")
````

````python
# not : inverser une condition
echantillon_valide = False

if not echantillon_valide:
    print("⚠️  Échantillon invalide — analyse annulée")
````

## 7. `if` dans une boucle `for` — la combinaison clé

En pratique, on combine presque toujours `for` et `if` pour analyser une liste de données.

````python
# 🧪 Biochimie : interpréter une série d'absorbances
absorbances = [0.05, 0.35, 0.78, 1.02, 1.45]

for i, A in enumerate(absorbances):
    if A < 0.1:
        print(f"Tube {i+1} | A = {A} → En dessous du seuil")
    elif A <= 0.9:
        print(f"Tube {i+1} | A = {A} → ✅ Zone linéaire")
    elif A <= 1.2:
        print(f"Tube {i+1} | A = {A} → ⚠️  Limite — diluer")
    else:
        print(f"Tube {i+1} | A = {A} → ❌ Hors gamme")
````

````python
# 🦠 Microbiologie : antibiogramme automatique
antibiotiques = ["Ampicilline", "Tétracycline", "Ciprofloxacine", "Érythromycine"]
diamètres     = [6, 22, 28, 15]
seuil_S       = 20   # ≥ 20 mm : Sensible
seuil_I       = 15   # 15-19 mm : Intermédiaire  / < 15 mm : Résistant

for i, atb in enumerate(antibiotiques):
    d = diamètres[i]
    if d >= seuil_S:
        print(f"✅  {atb:<18} {d} mm → Sensible")
    elif d >= seuil_I:
        print(f"⚠️   {atb:<18} {d} mm → Intermédiaire")
    else:
        print(f"❌  {atb:<18} {d} mm → Résistant")
````

````python
# 💊 Pharmacologie : doses patients
patients = ["Ali", "Sara", "Karim", "Lina"]
poids    = [70, 55, 90, 48]
dose_kg  = 5
dose_max = 400

for i in range(len(patients)):
    dose = poids[i] * dose_kg
    if dose > dose_max:
        print(f"⚠️  {patients[i]} ({poids[i]} kg) : dose TROP ÉLEVÉE ({dose} mg)")
    else:
        print(f"✅  {patients[i]} ({poids[i]} kg) : dose correcte ({dose} mg)")
````

````python
# ❌ ERREUR 1 : confondre = et ==
# if pH = 7:    → SyntaxError
# if pH == 7:   → ✅ correct

# ❌ ERREUR 2 : oublier les deux-points
# if pH > 7
#     print("basique")   → SyntaxError

# ❌ ERREUR 3 : mauvaise indentation
# if pH > 7:
# print("basique")   → IndentationError

# ❌ ERREUR 4 : elif après else
# if pH < 7:
#     print("acide")
# else:
#     print("autre")
# elif pH == 7:   → SyntaxError — else doit toujours être EN DERNIER

# ✅ Ordre correct : if → elif → elif → else
pH = 7.4
if pH < 7.0:
    print("Acide")
elif pH == 7.0:
    print("Neutre")
else:
    print("Basique")
````

## 📋 Résumé

```python
# Structure complète
if condition_1:       # si
    # cas 1
elif condition_2:     # sinon si
    # cas 2
elif condition_3:     # sinon si
    # cas 3
else:                 # sinon (tous les autres cas)
    # cas par défaut
```

| Mot-clé | Rôle | Position |
|---|---|---|
| `if` | Teste la première condition | Toujours **en premier** |
| `elif` | Teste une autre condition si la précédente est fausse | Après `if`, autant de fois que nécessaire |
| `else` | Capture tous les autres cas | Toujours **en dernier**, facultatif |
| `and` | Les deux conditions doivent être vraies | Dans la condition |
| `or` | Au moins une condition vraie suffit | Dans la condition |
| `not` | Inverse la condition | Dans la condition |

### Règle d'or
> L'ordre est toujours : `if` → `elif` → `elif` → `else`  
> Dès qu'une condition est vraie, Python exécute son bloc et **ignore tout le reste**.
