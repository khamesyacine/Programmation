## 1. Pourquoi a-t-on besoin d'une boucle ?

Imagine que tu travailles au laboratoire. Tu as **10 tubes** à analyser, et pour chaque tube tu dois afficher son numéro.

**Sans boucle**, tu écrirais :

````python
# ❌ Sans boucle : long, répétitif, impossible pour 1000 échantillons
print("Analyse tube 1")
print("Analyse tube 2")
print("Analyse tube 3")
# ... jusqu'à 10 ?
````

````python
# ✅ Avec une boucle for : 2 lignes suffisent pour 10 tubes (ou 1000 !)
for i in range(1, 11):
    print("Analyse tube", i)
````

---
## 2. La syntaxe — et pourquoi elle est écrite comme ça

```python
for variable in séquence:
    # code répété à chaque tour
```

| Élément | Rôle |
|---|---|
| `for` | Mot-clé qui démarre la boucle. Signifie *"pour chaque"* |
| `variable` | Prend une nouvelle valeur à chaque tour (`i`, `tube`, `patient`...) |
| `in` | Relie la variable à la séquence — *"parmi les éléments de"* |
| `séquence` | La liste de valeurs à parcourir |
| `:` | **Obligatoire** — annonce le début du bloc répété |
| **indentation** | 4 espaces — tout ce qui est décalé fait partie de la boucle |

````python
# Démonstration de l'indentation
for i in range(3):
    print("je suis DANS la boucle - tour", i)   # répété 3 fois
    print("moi aussi !")                          # répété 3 fois

print("je suis HORS de la boucle")               # affiché UNE seule fois
````

## 3. `range()` — comment ça marche vraiment

`range()` génère une suite de nombres entiers. Elle accepte 1, 2 ou 3 arguments :

| Écriture | Résultat | Usage |
|---|---|---|
| `range(n)` | 0, 1, ..., n-1 | Répéter N fois |
| `range(début, fin)` | début, ..., fin-1 | Numéroter des tubes/réplicats |
| `range(début, fin, pas)` | sauts réguliers | Gamme étalon à pas régulier |

> ⚠️ **La valeur de fin est toujours EXCLUE** — c'est une convention Python.

````python
# Par défaut, print() saute à la ligne après chaque affichage. C'est son comportement normal.
print(1)
print(2)
print(3)
````

````python
# Chaque print va à la ligne parce que par défaut end="\n" — \n est le caractère saut de ligne.
print(1, end="\n")
print(2, end="\n")
print(3, end="\n")
````

````python
# Chaque print va à la ligne parce que par défaut end=" " 
print(1, end=" ")
print(2, end=" ")
print(3, end=" ")
````

````python
# range(n) : de 0 à n-1
print("range(5) :")
for i in range(5):
    print(i, end=" ")
````

````python
# range(début, fin) : de début à fin-1
print("range(1, 6) :")
for i in range(1, 6):
    print(i, end=" ")
````

````python
# range(début, fin, pas) : utile pour une gamme étalon
print("range(0, 11, 2) — concentrations espacées :")
for i in range(0, 11, 2):
    print(i, end=" ")
````

## 4. Parcourir une liste directement

Au lieu de travailler avec des numéros, tu peux parcourir directement une liste de données réelles.

**Pourquoi `for atb in antibiotiques` plutôt que `for i in range(...)` ?**  
Parce que quand tu as déjà une liste de noms ou de valeurs, c'est plus lisible et plus direct. Le code se lit comme une phrase : *"pour chaque antibiotique dans ma liste"*.

````python
antibiotiques = ["Ampicilline", "Tétracycline", "Ciprofloxacine"]

for atb in antibiotiques:
    print("Test en cours :", atb)
````

## 5. `enumerate()` — quand tu veux l'indice ET la valeur

Parfois tu as besoin à la fois du **numéro du tour** et de la **valeur**.

`enumerate()` retourne automatiquement un couple `(indice, valeur)` à chaque tour.

````python
patients = ["Ali", "Sara", "Karim"]

for i, nom in enumerate(patients):
    print(f"Patient {i+1} : {nom}")
````

## 6. Exemples complets dans votre domaine

### 🧪 Biochimie — Gamme étalon (Loi de Beer-Lambert)

**Rappel :** A = ε × l × c

La boucle calcule l'absorbance pour chaque concentration sans répéter la formule.

````python
epsilon = 6500    # coefficient d'extinction molaire (L/mol/cm)
l = 1             # longueur du trajet optique (cm)

concentrations = [0.001, 0.002, 0.005, 0.010]  # mol/L

print("--- Gamme étalon ---")
for c in concentrations:
    A = epsilon * l * c
    print(f"c = {c} mol/L  →  Absorbance = {A:.3f}")
````

### 💊 Pharmacologie — Calcul de dose et détection de seuil toxique

On utilise `range(len(patients))` ici parce qu'on a besoin de l'indice `i` pour accéder **simultanément** aux deux listes `patients` et `poids`.

````python
patients = ["Ali", "Sara", "Karim", "Lina"]
poids    = [70, 55, 90, 48]     # kg
dose_kg  = 5                    # 5 mg/kg
dose_max = 400                  # seuil toxique en mg

print("--- Calcul des doses ---")
for i in range(len(patients)):
    dose = poids[i] * dose_kg
    if dose > dose_max:
        print(f"⚠️  {patients[i]} ({poids[i]} kg) : dose TROP ÉLEVÉE ({dose} mg)")
    else:
        print(f"✅  {patients[i]} ({poids[i]} kg) : dose correcte ({dose} mg)")
````

### 🦠 Microbiologie — Croissance bactérienne

On écrit `temps + 1` dans `range()` parce que la borne de fin est exclue — sans ce `+1`, la dernière heure ne serait jamais affichée.

````python
N0    = 100   # population initiale
temps = 8     # durée en heures

print("--- Croissance bactérienne (doublement par heure) ---")
for t in range(0, temps + 1):
    N = N0 * (2 ** t)
    print(f"t = {t}h  →  {N} bactéries")
````

### 🦠 Microbiologie — Lecture automatique d'antibiogramme

````python
antibiotiques = ["Ampicilline", "Tétracycline", "Ciprofloxacine", "Érythromycine"]
diamètres     = [6, 22, 28, 15]   # diamètres en mm
seuil         = 20                # seuil de sensibilité (mm)

print("--- Résultats de l'antibiogramme ---")
for i, atb in enumerate(antibiotiques):
    if diamètres[i] >= seuil:
        print(f"✅  {atb} : Sensible ({diamètres[i]} mm)")
    else:
        print(f"❌  {atb} : Résistant ({diamètres[i]} mm)")
````

### 📊 Calcul de moyenne sur des mesures répétées

Situation classique : tu as répété une mesure 5 fois (DO, pH, absorbance...).

````python
mesures = [0.452, 0.448, 0.461, 0.455, 0.450]

total = 0
for valeur in mesures:
    total = total + valeur

moyenne = total / len(mesures)
print(f"Mesures : {mesures}")
print(f"Moyenne des DO : {moyenne:.3f}")
````

## 📋 Résumé

| Tu veux... | Tu écris... |
|---|---|
| Répéter N fois | `for i in range(N):` |
| De 1 à N | `for i in range(1, N+1):` |
| Parcourir une liste | `for x in ma_liste:` |
| Indice + valeur | `for i, x in enumerate(ma_liste):` |
| Deux listes en parallèle | `for i in range(len(liste)):` |

---

> **Prochaine étape :** la boucle `while` — quand on répète non pas sur une liste, mais tant qu'une condition est vraie (ex : tant que la bactérie se multiplie, tant que la concentration dépasse le seuil...).
