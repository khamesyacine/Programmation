## 1. D'abord — tu connais déjà la boucle `for`

Avant de comprendre `while`, rappelons ce que fait `for` :

> La boucle `for` répète un code **un nombre de fois connu à l'avance**.

````python
# for : on sait qu'on a 5 tubes → on répète 5 fois
for i in range(1, 6):
    print("Analyse tube", i)
````

Mais parfois, **on ne sait pas à l'avance combien de fois répéter**.

Exemples :
- On dilue un échantillon **jusqu'à ce que** la concentration soit assez faible
- On attend **jusqu'à ce que** le pH atteigne 7.4
- On élimine un médicament **jusqu'à ce que** la dose soit nulle

Dans ces cas, on ne sait pas combien de tours seront nécessaires. On répète **tant qu'une condition est vraie**.

**C'est exactement ce que fait `while`.**

---
## 2. La différence entre `for` et `while`

| | `for` | `while` |
|---|---|---|
| **Quand l'utiliser** | Nombre de répétitions **connu** | Nombre de répétitions **inconnu** |
| **Logique** | *"répète 10 fois"* | *"répète tant que..."* |
| **Exemple** | Parcourir 5 tubes | Diluer jusqu'au seuil |

> **Règle simple :** si tu peux répondre *"combien de fois ?"* → `for`. Si tu réponds *"jusqu'à quand ?"* → `while`.

---
## 3. La syntaxe du `while` — mot par mot

```python
while condition:
    # code répété tant que la condition est vraie
```

Décomposons chaque élément :

| Élément | Rôle |
|---|---|
| `while` | Mot-clé qui démarre la boucle. Signifie **"tant que"** |
| `condition` | Une expression vraie (`True`) ou fausse (`False`) |
| `:` | Obligatoire — annonce le début du bloc répété |
| indentation (4 espaces) | Tout ce qui est décalé est répété |

Le fonctionnement est simple :
1. Python vérifie la condition
2. Si elle est **vraie** → il exécute le bloc, puis revient à l'étape 1
3. Si elle est **fausse** → il sort de la boucle et continue

---
## 4. Premier exemple — le plus simple possible

Comptons de 1 à 5 avec `while`.

````python
compteur = 1                  # on part de 1

while compteur <= 5:          # tant que compteur est ≤ 5
    print(compteur)           # affiche la valeur
    compteur = compteur + 1   # passe au suivant

print("Boucle terminée !")
````

Voici ce que Python fait à chaque tour :

| Tour | `compteur` | Condition `compteur <= 5` | Action |
|---|---|---|---|
| 1 | 1 | ✅ True | affiche 1, compteur devient 2 |
| 2 | 2 | ✅ True | affiche 2, compteur devient 3 |
| 3 | 3 | ✅ True | affiche 3, compteur devient 4 |
| 4 | 4 | ✅ True | affiche 4, compteur devient 5 |
| 5 | 5 | ✅ True | affiche 5, compteur devient 6 |
| 6 | 6 | ❌ False | **la boucle s'arrête** |

---
## 5. ⚠️ La règle la plus importante — modifier la variable !

Regarde ce que `compteur = compteur + 1` fait :

```
compteur = compteur + 1
    ↑           ↑
 nouvelle    ancienne
  valeur      valeur
```

**C'est OBLIGATOIRE.** Si on oublie cette ligne, `compteur` reste toujours à 1, la condition est toujours vraie, et la boucle ne s'arrête jamais.

On appelle ça une **boucle infinie** — c'est l'erreur la plus dangereuse avec `while`.

````python
# ❌ NE PAS EXÉCUTER — boucle infinie !
# compteur = 1
# while compteur <= 5:
#     print(compteur)
#     # OUBLI : pas de compteur = compteur + 1
#     # compteur reste à 1 pour toujours → boucle infinie !

# ✅ Toujours modifier la variable à chaque tour
compteur = 1
while compteur <= 5:
    print(compteur)
    compteur = compteur + 1   # ← indispensable
````

### Le raccourci `+=`

`compteur = compteur + 1` s'écrit aussi `compteur += 1` — c'est exactement la même chose, juste plus court.

| Longue écriture | Raccourci |
|---|---|
| `compteur = compteur + 1` | `compteur += 1` |
| `compteur = compteur - 1` | `compteur -= 1` |
| `compteur = compteur * 2` | `compteur *= 2` |
| `compteur = compteur / 2` | `compteur /= 2` |

````python
# Les deux écritures donnent le même résultat
compteur = 1
while compteur <= 5:
    print(compteur, end=" ")
    compteur += 1   # raccourci de : compteur = compteur + 1
````

## 6. Les opérateurs de condition

La condition du `while` utilise des opérateurs de comparaison :

| Opérateur | Signification | Exemple |
|---|---|---|
| `<` | strictement inférieur | `concentration < 0.5` |
| `<=` | inférieur ou égal | `dose <= 400` |
| `>` | strictement supérieur | `N > 1000` |
| `>=` | supérieur ou égal | `pH >= 7.4` |
| `==` | égal | `résultat == 0` |
| `!=` | différent | `statut != "mort"` |

> ⚠️ Ne jamais confondre `=` (donner une valeur) et `==` (comparer deux valeurs)

````python
# Démonstration : ajuster le pH progressivement
pH = 6.0

while pH < 7.4:               # tant que le pH est trop acide
    print(f"pH = {pH:.1f} → ajout de tampon...")
    pH += 0.3                 # on ajoute du tampon

print(f"pH = {pH:.1f} → pH physiologique atteint ✅")
````

## 7. Exemples du domaine

### 🧪 Biochimie — Dilutions successives

On dilue de moitié à chaque fois, jusqu'à passer sous le seuil de détection.  
On ne sait pas combien de dilutions seront nécessaires → `while`.

````python
concentration = 1.0      # mg/mL — concentration initiale
seuil         = 0.05     # mg/mL — seuil de détection
etape         = 1

print(f"Départ : {concentration:.4f} mg/mL")
print("---")

while concentration > seuil:      # tant que trop concentré
    concentration /= 2            # divise par 2
    print(f"Dilution {etape} : {concentration:.4f} mg/mL")
    etape += 1

print("---")
print(f"Seuil atteint après {etape - 1} dilutions ✅")
````

### 💊 Pharmacologie — Élimination d'un médicament

Un médicament est éliminé à 15% par heure.  
On surveille combien d'heures il reste au-dessus du seuil thérapeutique.

````python
dose    = 500      # mg dans le sang
seuil   = 50       # mg — en dessous : plus d'effet
heure   = 0

print(f"t = {heure}h  →  {dose:.1f} mg")

while dose >= seuil:               # tant qu'efficace
    dose  *= (1 - 0.15)            # 15% éliminé par heure
    heure += 1
    print(f"t = {heure}h  →  {dose:.1f} mg")

print(f"\n⚠️  Médicament sous le seuil à t = {heure}h")
print(f"→  Renouveler la dose après {heure} heures")
````

### 🦠 Microbiologie — Croissance jusqu'à saturation

Dans un milieu limité, les bactéries croissent jusqu'à une capacité maximale K.  
On ne sait pas quand la saturation sera atteinte → `while`.

````python
N     = 100       # population initiale
K     = 10000     # capacité maximale du milieu
r     = 0.5       # taux de croissance
heure = 0

print(f"t = {heure}h  →  N = {N:.0f} bactéries")

while N < K * 0.99:               # tant que pas saturé à 99%
    N     += r * N * (1 - N/K)   # modèle logistique
    heure += 1
    print(f"t = {heure}h  →  N = {N:.0f} bactéries")

print(f"\n🔬 Saturation atteinte à t = {heure}h")
````

## 8. `break` — sortir de la boucle en urgence

`break` dit à Python : *"arrête immédiatement la boucle, même si la condition est encore vraie"*.

C'est utile quand un événement imprévu doit stopper l'expérience.

````python
# On surveille la température d'un réacteur
# Si elle dépasse 42°C → arrêt d'urgence
temperatures = [36.5, 37.2, 38.0, 39.5, 43.2, 38.5]
i = 0

while i < len(temperatures):
    t = temperatures[i]
    print(f"Mesure {i+1} : {t}°C")

    if t > 42:
        print("🚨 TEMPÉRATURE CRITIQUE — Arrêt d'urgence !")
        break                  # sort immédiatement de la boucle

    i += 1
````

## 9. `continue` — sauter un tour

`continue` dit à Python : *"ignore le reste de ce tour et passe directement au suivant"*.

Utile pour ignorer les valeurs aberrantes ou les échantillons invalides.

````python
# On analyse des absorbances
# Les valeurs négatives sont des erreurs de mesure → on les ignore
absorbances = [0.45, 0.52, -0.03, 0.48, -0.01, 0.50]
i = 0

while i < len(absorbances):
    a  = absorbances[i]
    i += 1

    if a < 0:
        print(f"⚠️  Valeur aberrante ignorée : {a}")
        continue               # saute le print ci-dessous

    print(f"✅  Absorbance valide : {a}")
````

## 10. `for` ou `while` — comment choisir ?

Pose-toi toujours cette question avant d'écrire ta boucle :

| Question | Réponse | Utilise |
|---|---|---|
| Je connais le nombre exact de répétitions ? | Oui → 10 tubes, 5 patients | `for` |
| J'ai une liste à parcourir ? | Oui | `for` |
| Je répète jusqu'à ce qu'une valeur change ? | Oui | `while` |
| Je ne sais pas combien de tours seront nécessaires ? | Oui | `while` |

````python
# ✅ for — je sais que j'ai 5 tubes
print("--- for ---")
for i in range(1, 6):
    print("Tube", i)

print()

# ✅ while — je sais pas combien de dilutions il faudra
print("--- while ---")
c = 1.0
while c > 0.1:
    c /= 2
    print(f"c = {c:.3f} mg/mL")
````

## 11. Les erreurs les plus fréquentes

````python
# ❌ ERREUR 1 : oublier les deux-points
# while compteur < 5
#     print(compteur)
# → SyntaxError

# ❌ ERREUR 2 : oublier de modifier la variable → boucle infinie
# compteur = 0
# while compteur < 5:
#     print(compteur)
#     # pas de compteur += 1 → boucle infinie !

# ❌ ERREUR 3 : confondre = et ==
# while pH = 7:    → SyntaxError
# while pH == 7:   → ✅ correct

# ✅ Code correct
compteur = 0
while compteur < 5:
    print(compteur, end=" ")
    compteur += 1
````

## 📋 Résumé

```python
# Structure de base
while condition:
    # code répété
    # ⚠️ toujours modifier la variable de condition !
```

| Élément | Rôle |
|---|---|

| `condition` | Expression vraie ou fausse |

### Règle d'or
